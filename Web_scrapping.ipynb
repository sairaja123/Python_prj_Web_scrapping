```python
# Import Neccessary Libraries
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_Canada'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

soup
```

```python
soup.find('table')
```

```python
soup.find_all('table')
```

```python
soup.find_all('table')[1]
```

```python
# Select the Required table
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
fortune_titles = table.find_all('th')
```

```python
fortune_titles
```

```python
fortune_table_titles = [title.text.strip() for title in fortune_titles]

print(fortune_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = fortune_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

# Data Cleaning

```python
import pandas as pd


file_path = 'Traded_Companies.csv'

df = pd.read_csv(file_path)

# Display the DataFrame to verify the data has been loaded
print(df)
```

```python


import pandas as pd


# Convert string columns to numeric
numeric_cols = ['Profit(billions US$)', 'Rank', 'Forbes 2000 rank', 'Revenue(billions US$)', 'Assets(billions US$)', 'Value(billions US$)']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Clean the values in the columns (for example, by taking absolute values and rounding)
df[numeric_cols] = df[numeric_cols].abs().round(1)

# Display the cleaned DataFrame
print(df)
```

```python
df.to_csv(r'C:\Users\ravul\Downloads\St. clairs\python\Python project\prj\Traded_Companies.csv', index = False)
```
